```python


# Define the MQTT broker and port
broker = "broker.hivemq.com"
port = 1883

# Callback function when the message is published
def on_publish(client, userdata, mid):
    print(f"Message published with mid: {mid}")

# Create an MQTT client instance
client = mqtt.Client()

# Bind the on_publish callback
client.on_publish = on_publish

# Connect to the MQTT broker
client.connect(broker, port, 60)

# Start the MQTT client loop in a separate thread
client.loop_start()

try:
    while True:
        # Generate random sound level between 30 and 100 dB
        sound_level = round(random.uniform(30, 100), 2)  # Rounded to 2 decimal places
        topic = "kelas_iot/sound"
        message = str(sound_level)
        result = client.publish(topic, message)
        # Print the published message
        print(f"Published {message} dB")
        time.sleep(2)
except KeyboardInterrupt:
    print("Stopping the MQTT publisher...")
finally:
    # Stop the MQTT client loop and disconnect
    client.loop_stop()
    client.disconnect()
```
